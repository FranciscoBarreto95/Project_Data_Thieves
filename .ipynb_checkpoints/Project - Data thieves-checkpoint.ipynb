```python
#Used the google news api and newspapper3k to scrape headlines
from GoogleNews import GoogleNews
#pip install newspaper3k - need installed 
from newspaper import Article
```

```python
#retrieve the headlines from googlenews for each ticker 
list_of_news=[]
for ticker in ticker_names:
    googlenews = GoogleNews()
    googlenews.search(ticker)
    result = googlenews.result()
    for news in result:
        news["ticker"] = ticker
        list_of_news.append(result)

df = pd.DataFrame(list_of_news[0])
for i in range(1,len(list_of_news)):
    df = pd.concat([df,pd.DataFrame(list_of_news[i])])
```

```python
News_df=Finance_news[["ticker","title","desc"]]
```

```python
News_df
```

```python
#sentimental analysis - will produce a sentimental analysis of each title head line
#pip install textblob
from textblob import TextBlob
```

```python
sentiment_list = []
for i in range(0,len(News_df["title"])):
    blob = TextBlob(News_df["title"][i])
    sentiment = blob.sentiment.polarity #-1 to 1
    sentiment_list.append(sentiment)

sentiment_list
```

```python
News_df["Sentiment"]= sentiment_list
```

```python
News_df
```

```python
#I am going to agregate this data per ticker and by the average of the sentimental value 
```

```python
Ticker_sentiment_df=News_df.groupby("ticker").agg({"Sentiment":"mean"})
Ticker_sentiment_df
```

```python
Trading_view_clean_df=trading_index[["Ticker_names","Last","Chg %","Chg","Technical Rating"]]
Trading_view_clean_df
```

```python
dfinal = Trading_view_clean_df.merge(Ticker_sentiment_df, how='inner', left_on='Ticker_names', right_on='ticker')
```

```python
dfinal
```

```python
#used this to check what ratings were used to create the function below
Technical_Rating_list=list(pd.unique(dfinal["Technical Rating"]))
Technical_Rating_list
```

```python
#going to create a Tecnical Rating score in order to correlate and better understand the data 
#Strong Sell -> -2
#Sell -> -1
#Neutral - > 0
#Buy -> 1

def score_rating(row):
    if row =='Strong Sell':
        return (-2)
    elif row =='Sell':
           return (-1)
    elif row =='Neutral':
        return (0)
    elif row =='Buy':
         return(1)

```

```python
#used to see  if the function worked
#print(score_rating('Strong Sell'))
```

```python
score_rating_l =[]
for i in list(dfinal["Technical Rating"]):
    score_rating_l.append(score_rating(i))
```

```python
score_rating_l
```

```python
#added the score_rating to dataframe
dfinal["score_rating"] = score_rating_l
```
